## 1. Introduction to Data Collection

### Importance of Data Collection in Machine Learning and Data Science

Data collection is the foundation of any Machine Learning (ML) or Data Science (DS) project. The quality, quantity, and relevance of the data you gather directly impact the performance and accuracy of the models you build. In ML and DS, data acts as the raw material that fuels algorithms, helping them to learn patterns, make predictions, and generate insights. Without a robust and well-organized data collection process, even the most sophisticated models can produce unreliable results.

Effective data collection involves gathering data that is representative of the problem you're trying to solve, ensuring that it covers all relevant aspects and is free from significant biases. It is also crucial to collect data that is clean, well-labeled, and organized, as poor data quality can lead to increased complexity in data preprocessing and can significantly hinder model performance.

### Overview of Web-Based Data Sources

Web-based data sources have become increasingly popular in recent years due to the vast amount of information available online. These sources offer a treasure trove of data that can be harnessed for various applications in ML and DS. Common web-based data sources include:

1. **APIs (Application Programming Interfaces):** Many websites and services provide APIs that allow users to access structured data programmatically. Examples include social media APIs (like Twitter's API), financial data APIs (like Alpha Vantage), and weather data APIs (like OpenWeatherMap).

2. **Web Scraping:** This technique involves extracting data directly from websites. Using libraries like BeautifulSoup or Scrapy, data scientists can scrape data from web pages and convert it into a structured format for analysis.

3. **Open Data Repositories:** Numerous organizations and governments provide free access to datasets through open data repositories. Examples include Kaggle, UCI Machine Learning Repository, and government portals like data.gov.

4. **Crowdsourcing Platforms:** Platforms like Amazon Mechanical Turk or data donation initiatives allow data scientists to collect data directly from individuals. This can be particularly useful for gathering labeled datasets for supervised learning tasks.

### Ethical Considerations in Data Collection

When collecting data, especially from web-based sources, it's essential to consider the ethical implications of your actions. Ethical data collection practices not only protect the rights of individuals but also ensure compliance with legal standards and maintain the integrity of your research. Key ethical considerations include:

1. **Consent:** Ensure that individuals whose data you are collecting have given their explicit consent. This is particularly important when collecting data that can be traced back to specific individuals, such as personal information or social media activity.

2. **Privacy:** Protect the privacy of individuals by anonymizing data whenever possible. Avoid collecting personally identifiable information (PII) unless absolutely necessary, and ensure that any sensitive data is stored securely.

3. **Terms of Service:** Always adhere to the terms of service of the websites or platforms from which you are collecting data. Violating these terms can lead to legal repercussions and damage your reputation.

4. **Bias and Fairness:** Be mindful of biases that may be present in the data you collect. Strive to collect data that is representative of the entire population or domain you are studying to avoid perpetuating or amplifying existing biases in your ML models.

In summary, data collection is a critical step in the ML and DS pipeline. By understanding the importance of data collection, utilizing web-based data sources effectively, and adhering to ethical standards, you can ensure that the data you collect is both useful and responsibly gathered.

## 2. Methods of Collecting Data from Web-Based Sources

When working on Machine Learning (ML) and Data Science (DS) projects, collecting data from web-based sources is a common and often necessary task. There are several methods for gathering this data, each with its own advantages and challenges. Understanding these methods and their respective pros and cons can help you choose the most appropriate approach for your specific needs.

### Manual Data Collection: Pros and Cons

**Manual data collection** involves the process of manually searching for, selecting, and recording data from web-based sources. This method is often used when working with small datasets or when the data is not easily accessible through automated means.

**Pros:**
- **Precision and Control:** Manual data collection allows for precise selection and filtering of data. You can ensure that only relevant and high-quality data is included in your dataset.
- **Flexibility:** You can adapt quickly to changes or unexpected challenges, such as accessing data that requires a human judgment or dealing with non-standardized content.
- **No Coding Required:** This method does not require programming skills or specialized tools, making it accessible to those who are not familiar with coding or automation.

**Cons:**
- **Time-Consuming:** Manual data collection is often slow and labor-intensive, especially when dealing with large volumes of data.
- **Inconsistency:** The manual process can lead to inconsistencies or errors due to human fatigue or oversight.
- **Limited Scalability:** This approach does not scale well for large datasets, making it impractical for projects that require vast amounts of data.

### Automated Data Collection Techniques

For larger projects or when efficiency and scalability are important, **automated data collection** techniques are generally preferred. These methods leverage software tools to gather data from web-based sources quickly and consistently.

**1. Web Scraping**

**Web scraping** involves using software tools to automatically extract data from websites. This technique is especially useful when dealing with structured data embedded in HTML, such as tables, lists, or articles.

**Pros:**
- **Efficiency:** Web scraping can extract large amounts of data in a relatively short period, far more quickly than manual collection.
- **Automation:** Once set up, web scraping scripts can run with minimal human intervention, allowing for the automated collection of real-time or updated data.
- **Customization:** Scraping tools can be customized to target specific elements of a webpage, allowing for tailored data collection based on project needs.

**Cons:**
- **Legal and Ethical Concerns:** Web scraping can violate a website's terms of service, and scraping certain types of data without permission may lead to legal issues.
- **Complexity:** Scraping can be technically complex, especially when dealing with dynamic web pages, CAPTCHAs, or anti-scraping mechanisms.
- **Maintenance:** Web scraping scripts may break if the structure of the targeted website changes, requiring regular updates and maintenance.

**2. APIs (Application Programming Interfaces)**

**APIs** are interfaces provided by websites or services that allow users to programmatically access data in a structured and often real-time manner. Many modern web services offer APIs for developers to integrate and interact with their platforms.

**Pros:**
- **Structured Data:** APIs typically return data in structured formats like JSON or XML, making it easy to integrate and analyze.
- **Reliability:** APIs are designed for data exchange, so they are generally reliable and well-documented, with clear guidelines on how to use them.
- **Real-Time Access:** APIs often provide real-time data, which is valuable for applications that require up-to-date information.

**Cons:**
- **Rate Limits:** Many APIs impose rate limits, restricting the number of requests you can make in a given time period. This can be a limitation for projects requiring large amounts of data.
- **Authentication and Access:** Some APIs require authentication, which may involve obtaining and managing API keys or tokens. Additionally, some APIs are not free and may require payment for access.
- **Limited Availability:** Not all websites or services offer APIs, and even when they do, the available data may be limited to what the API provider chooses to expose.

**3. Public Datasets and Repositories**

**Public datasets and repositories** are collections of data made freely available by governments, organizations, or research institutions. These datasets are often pre-cleaned and ready for analysis, covering a wide range of topics such as finance, healthcare, environment, and social sciences.

**Pros:**
- **Accessibility:** Public datasets are typically free to use and widely available, making them an excellent resource for data-driven projects.
- **Variety:** There is a vast array of public datasets available, covering many different domains and research areas.
- **Pre-Processed Data:** Many public datasets are already cleaned and formatted, reducing the time and effort required for data preprocessing.

**Cons:**
- **Limited Customization:** You can only work with the data that is provided, which may not fully meet your specific needs or may lack the granularity required for certain analyses.
- **Outdated Data:** Some public datasets may not be regularly updated, leading to the use of stale data that does not reflect the current situation.
- **Inconsistencies:** While many public datasets are well-maintained, some may contain inconsistencies or errors that require additional cleaning and validation.

### Collecting data from web-based sources

With those general caveats in mind, let's dive a bit more deeply into the specific case of gather data from an web-based source, which is one of the more common forms of querying data.  

The first step of collecting web-based data is to issue a request for this data via some protocol: HTTP (HyperText Transfer Protocol) or HTTPS (the secure version).  For our purposes, we're going to use the [requests](http://docs.python-requests.org/en/master/) library in Python.

Let's see how this works with some code.  The following code will load data from the Wiki webpage:

````python
import requests

url = "https://en.wikipedia.org/wiki/Data_science"

response = requests.get(url)

print("Status Code:", response.status_code)
print("Headers:", response.headers)
````

````python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers)

print("Status Code:", response.status_code)
print("Headers:", response.headers)
````

This code issues an "HTTP GET" request to load the content of the paper, and returns it in the `response` object.  The `status_code` field contains the code, which indicates a successful (or not) query, and the `headers` field contains meta-information about the page (in this case, you could see, for instance, that despite the URL, we're actually hosting this page on github).  If you want to see the actual content of the page, you can use the `response.content` or `response.text` fields, as below.

````python
print(response.text[:480])
````

### RESTful APIs

While parsing data in HTML (the format returned by these web queries) is sometimes a necessity, HTML is meant as a format for displaying pages visually, not as the most efficient manner for encoding data.  Fortunately, a fair number of web-based data services you will use in practice employ something called REST (Representational State Transfer, but no one uses this term) APIs.  We won't go into detail about REST APIs, but there are a few main feature that are important for our purposes:

1. You call REST APIs using standard HTTP commands: GET, POST, DELETE, PUT.  You will probably see GET and POST used most frequently.
2. REST servers don't store state.  This means that each time you issue a request, you need to include all relevant information like your account key, etc.
3. REST calls will usually return information in a nice format, typically JSON (more on this later).  The `requests` library will automatically parse it to return a Python dictionary with the relevant data.

## 3.Introduction to Web Scraping

### Definition and Use Cases

**Web scraping** is the process of automatically extracting data from websites. It involves writing code to send HTTP requests to web pages, retrieving their content, and then parsing that content to extract specific pieces of data. Web scraping is particularly useful for gathering large amounts of data from websites where an API (Application Programming Interface) is not available or is too limited for your needs.

**Use Cases of Web Scraping:**

1. **Data Aggregation:** Web scraping is commonly used to aggregate data from multiple sources. For example, price comparison websites use scraping to collect pricing information from various online retailers to provide users with up-to-date comparisons.
   
2. **Market Research:** Companies often scrape data from competitors’ websites to analyze their offerings, pricing strategies, and customer reviews. This information helps in making informed business decisions.
   
3. **Academic Research:** Researchers use web scraping to collect data from various online sources, such as social media platforms, news sites, and forums, for sentiment analysis, trend analysis, and other research purposes.
   
4. **Job Listings:** Scraping job boards and company websites can help in creating a centralized database of job listings, which can be used for job search platforms or for analyzing job market trends.
   
5. **Real Estate Listings:** Real estate websites use web scraping to gather property listings from multiple sources to provide comprehensive databases of available properties.

### Overview of Web Scraping Tools and Libraries

There are several tools and libraries available for web scraping, each with its strengths and suited for different levels of complexity and requirements. Below are two popular tools used in Python:

#### BeautifulSoup

**BeautifulSoup** is a Python library used for parsing HTML and XML documents. It creates a parse tree that can be used to extract data from HTML, which is particularly useful for web scraping. BeautifulSoup is known for its simplicity and ease of use, making it an excellent choice for beginners or when working on small to medium-sized projects.

**Key Features of BeautifulSoup:**

- **Easy to Learn and Use:** BeautifulSoup provides a simple API for navigating and searching the parse tree, which makes it accessible even to those who are new to web scraping.
- **Handles Different Formats:** BeautifulSoup can parse HTML, XML, and other markup languages. It can handle poorly formatted or invalid markup, which is common on the web.
- **Integrates with Requests:** BeautifulSoup works seamlessly with the `requests` library, allowing you to send HTTP requests and then parse the resulting HTML with ease.

**Example Use Case:**
BeautifulSoup is ideal for projects where you need to extract data from a single or a few web pages. For example, you might use it to scrape the titles and prices of books from an online bookstore.

#### Scrapy

**Scrapy** is a more advanced and robust web scraping framework in Python. It is designed for large-scale scraping projects where you need to crawl and scrape multiple pages or even entire websites. Scrapy provides tools to handle requests, follow links, process data, and store the scraped data, all in an efficient and scalable manner.

**Key Features of Scrapy:**

- **Crawling:** Scrapy can follow links on a web page and scrape data from multiple pages, making it suitable for crawling entire websites.
- **Speed and Efficiency:** Scrapy is optimized for speed, using asynchronous requests to handle multiple HTTP requests simultaneously, which significantly speeds up the scraping process.
- **Built-In Data Storage:** Scrapy allows you to store scraped data in various formats such as JSON, CSV, or directly into databases, making it easy to manage and analyze large datasets.
- **Extensibility:** Scrapy is highly customizable, allowing you to write custom middleware and pipelines to process the data as it’s being scraped.

**Example Use Case:**
Scrapy is perfect for scraping large-scale data, such as collecting all product listings from an e-commerce website or gathering data from multiple news websites to create a news aggregation platform.

### Legal and Ethical Issues in Web Scraping

While web scraping is a powerful tool, it is essential to consider the legal and ethical implications of this practice. Not all websites allow scraping, and some explicitly prohibit it in their terms of service. Understanding these issues is crucial to avoiding legal trouble and ensuring that your scraping activities are ethical.

**Legal Issues:**

- **Terms of Service Violations:** Many websites include clauses in their terms of service that prohibit scraping. Ignoring these clauses can result in legal action, including cease and desist orders or lawsuits.
- **Intellectual Property:** The content on websites is often protected by copyright laws. Scraping and redistributing this content without permission can be a violation of intellectual property rights.
- **Data Privacy:** If you are scraping websites that contain personal data, such as social media platforms, you must be cautious about violating privacy laws like GDPR (General Data Protection Regulation) in Europe. Scraping personal data without proper consent can lead to severe penalties.

**Ethical Issues:**

- **Respect for Website Resources:** Scraping can put a strain on the server resources of the target website, especially if done aggressively or without proper rate limiting. This can slow down the website for other users or even cause it to crash.
- **User Consent:** If scraping involves collecting data about individuals, such as posts or comments on social media, it is important to consider whether those individuals have consented to have their data collected and used in the manner you intend.
- **Transparency:** It’s generally good practice to be transparent about your scraping activities, especially if you are scraping data for commercial purposes. Informing the website owner or including a clear description of your intentions can help maintain trust and avoid conflicts.

In conclusion, web scraping is a valuable technique for data collection, but it must be approached with care. Choosing the right tool, understanding the legal landscape, and following ethical guidelines are all essential steps to ensure that your web scraping projects are both effective and responsible.

**Code Example: Scraping the Wikipedia Page on Data Science with BeautifulSoup**

````python
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page about Data Science
url = 'https://wiki.python.org/python/FrontPage.html'

# Send a GET request to the URL
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract the title of the page
    title = soup.find('h1', {'id': 'The_Python_Wiki'}).text
    print(f"Page Title: {title}\n")

    # Extract the main content of the page
    content = soup.find('div', {'id': 'content'})

    # Extract all the paragraphs in the main content
    paragraphs = content.find_all('p')

    # Print the first few paragraphs of the page
    for i, para in enumerate(paragraphs[:5], start=1):
        print(f"Paragraph {i}:\n{para.text}\n")

    # Extract and print all the section headings
    headings = content.find_all(['h2', 'h3'])
    print("Section Headings:")
    for heading in headings:
        print(heading.text.strip())

else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
````

#### **Explanation of the Code:**
- **Step 1: Sending a Request** - The script sends a GET request to the Wikipedia page on Data Science using the `requests` library.
- **Step 2: Parsing the HTML** - The page content is parsed using BeautifulSoup with the `html.parser`.
- **Step 3: Extracting the Title** - The script finds the main title of the Wikipedia page using the `<h1>` tag with `id='firstHeading'`.
- **Step 4: Extracting Content** - The main content of the page is located within the `<div>` tag with `class='mw-parser-output'`.
- **Step 5: Extracting Paragraphs** - The script extracts paragraphs (`<p>` tags) from the main content, and prints the first few paragraphs.

### XML/HTML

Last, another format you will likely encoder are XML/HTML documents.  XML files contain hierarchical content delineated by tags, like the following:

XML contains "open" tags denoted by brackets, like `<tag>`, which are then closed by a corresponding "close" tag `</tag>`.  The tags can be nested, and have optional attributes, of the form `attribute_name="attribute_value"`.  Finally, there are "open/close" tags that don't have any included content (except perhaps attributes), denoted by `<openclosetag/>`.

HTML, the standard for describing web pages, may seem syntactically similar to XML, but it is difficult to parse properly (open tags may not have closed tags, etc).  Generally speaking, HTML was developed to display content for the web, not to organize data, so a lot of invalid structure (like the aforementioned open without close) became standard simply because people frequently did this in practice, and so the data format evolved.  There are many libraries for Python, but a particularly nice one is [BeautifulSoup](https://www.crummy.com/software/BeautifulSoup/).  Beautiful soup was actually written for parsing HTML (it is a common tool for scraping web pages), but it works just as well for the more-structured XML format.  

````python
from bs4 import BeautifulSoup

root = BeautifulSoup("""
<tag attribute="value">
    <subtag>
        Some content for the subtag
    </subtag>
    <openclosetag attribute="value2"/>
    <subtag>
        Second one
    </subtag>
</tag>
""", "lxml-xml")

print(root, "\n")
print(root.tag.subtag, "\n")
print(root.tag.openclosetag.attrs)
````

The `BeautifulSoup()` call creates the object to parse, where the second argument specifies the parser ("lxml-xml" indicates that it is actually XML data, whereas "lxml" is the more common parser for parsing HTML files).  As illustrated above, when the hierarchical layout of the data is fairly simple, here a "tag" followed by a "subtag" (by default this will return the first such tag), or an "openclosetag", you can access the various parts of the hierarchy simply by a structure-like layout of the BeautifulSoup object.  Where this gets trickier is when there are multiple tags with the same name as the hierarchy level, as there is with the two "subtag" tags.  Above.  In this case, you can use the `find_all` function, which returns a list of all the subtags.

````python
print(root.tag.find_all("subtag"))
````

The nice thing about the `find_all` function is that you can call it at previous levels in the tree, and it will recurse down the whole document.  So we could have just as easily done.

````python
print(root.find_all("subtag"))
````

Let's end with a slightly more complex example, that looks through the Wiki page to get some content.  Again, the details aren't important here, but by playing around with these calls you'll get a sense of how to extract information from web pages or from XML documents in general.

### Parsing XML Files Using `xml.etree.ElementTree` and `lxml`

**XML (eXtensible Markup Language)** is a markup language used to encode documents in a format that is both human-readable and machine-readable. XML is commonly used for representing hierarchical data structures, such as configuration files, data interchange formats, and more.

**Using `xml.etree.ElementTree`:**
`xml.etree.ElementTree` is a part of Python’s standard library and provides a simple and efficient way to parse and create XML data. It is suitable for basic XML parsing tasks.

**Example: Parsing XML with `xml.etree.ElementTree`:**


````python
import xml.etree.ElementTree as ET

# Parse the XML file
tree = ET.parse(DATA_DIR / 'data.xml')
root = tree.getroot()

# Iterate over the elements
for child in root:
    print(child.tag, child.attrib)
````


**Using `lxml`:**
`lxml` is a more powerful and flexible library for parsing XML and HTML. It builds on `xml.etree.ElementTree` but offers additional features such as better performance, support for XPath and XSLT, and more robust handling of XML structures.

**Example: Parsing XML with `lxml`:**

````python
from lxml import etree

# Parse the XML file
tree = etree.parse(str(DATA_DIR / 'data.xml'))
root = tree.getroot()

# Use XPath to find specific elements (e.g., all employee names)
names = root.xpath('//name')
print("Employee names:")
for name in names:
    print(f"  - {name.text}")

# Find all employees
employees = root.xpath('//employee')
print(f"\nTotal employees: {len(employees)}")
````

**Key Differences:**
- `xml.etree.ElementTree` is ideal for simple XML parsing tasks where you need a lightweight and easy-to-use solution.

## 6. Regular expressions

The last tool we're going to consider in these notes are regular expressions.  Regular expressions are invaluable when parsing any type of unstructured data, if you're trying to quickly find or extract some text from a long string, and even if you're writing a more complex parser.  In general, regular expressions let us find and match portions of text using a simple syntax (by some definition).  

### Finding

Let's start with the most basic example, that simply searches text for some sting.  In this case, the text we are searching is _"This course will introduce the basics of data science"_, and the string we are searching for is _"data science"_.  This is done with the following code:

````python
import re
text = "This course will introduce the basics of data science"
match = re.search(r"data science", text)
print(match.start())
````

The important element here is the `re.search(r"data science", text)` call.  It searches `text` for the string "data science" and returns a regular expression "match" object that contains information about where this match was found: for instance, we can find the character index (in `text`) where the match is found, using the `match.start()` call.  In addition to the search call, there are two or three more regular expression matching commands you may find useful:

- `re.match()`: Match the regular expression starting at the _beginning_ of the text string
- `re.finditer()`: Find all matches in the text, returning a iterator over match objects
- `re.findall()`: Find all matches in the text, returning a list of the matched text only (not a match object)

For example, the following code would return `None`, since there is no match to _"data science"_ at the beginning of the string:

````python
match = re.match(r"data science", text)
print(match)
````

Similarly, we could use `re.finditer()` to list the location of all the 'i' characters in the string:

````python
for match in re.finditer(r"i", text):
    print(match.start())
````

On the other hand, `re.findall()` just returns a list of the matched strings, with no additional info such as where they occurred:

````python
re.findall(r"i", text)
````

This last call may not seem particularly useful, but especially when you use more complex matching expressions, this last call can still be of some use.  Finally, you can also "compile" a regular expression and then make all the same calls on this compiled object, as in the following:

````python
regex = re.compile(r"data science")
regex.search(text)
````

However, given that Python will actually compile expressions on the fly anyways, don't expect a big speed benefit from using the above; whether to use `re.compile()` separately is more of a personal preference than anything else.

### Matching multiple potential strings

While using regular expressions to search for a string within a long piece of text may be a handy tools, the real power of regular expressions comes from the ability to match multiple potential strings with a single regular expression.  This is where the syntax of regular expressions gets nasty, but here are some of the more common rules:

- Any character (except special characters, `".$*+?{}\[]|()` ), just matches itself.  I.e., the character `a` just matches the character `a`.  This is actually what we used previously, where each character in the `r"data science"` regular expression was just looking to match that exact character.
- Putting a group of characters within brackets `[abc]` will match any of the characters `a`, `b`, or `c`. You can also use ranges within these brackets, so that `[a-z]` matches any lower case letter.
- Putting a caret within the bracket matches anything _but_ these characters, i.e., `[^abc]` matches any character _except_ `a`, `b`, or `c`.
- The special character `\d` will match any digit, i.e. `[0-9]`
- The special character `\w` will match any alphanumeric character plus the underscore; i.e., it is equivalent to `[a-zA-Z0-9_]`.
- The special character `\s` will match whitespace, any of `[ \t\n\r\f\v]` (a space, tab, and various newline characters).
- The special character `.` (the period) matches _any_ character.  In their original versions, regular expressions were often applies line-by-line to a file, so by default `.` will _not_ match the newline character.  If you want it to match newlines, you pass `re.DOTALL` to the "flags" argument of the various regular expression calls.

As an example, the following regular expression will match "data science" regardless of the capitalization, and with any type of space between the two words.

````python
print(re.search(r"[Dd]ata\s[Ss]cience", text))
````

Note that now the match objects also now include what was the particular text that matched the expression (which could be one of any number of possibilities now).  This is why calls like `re.findall` are still useful even if they only return the matched expression.

The second important feature of regular expressions it the ability to match multiple _occurences_ of a character.  The most important rules here are:

- To match `a` exactly once, use `a`.
- To match `a` zero or one times, use `a?`.
- To match `a` zero or more times, use `a*`.
- To match `a` one or more times, use `a+`.
- And finally, to match `a` exactly n times, use `a{n}`.

These rules can of course be combined with the rules to match potentially very complicated expressions.  For instance, if we want to match the text "*something* science" where *something* is any alphanumeric character, and there can be any number of spaces of any kind between *something* and the word "science", we could use the expression `r"\w+\s+science"`.

````python
print(re.match("\w+\s+science", "data science"))
print(re.match("\w+\s+science", "life science"))
print(re.match("\w+\s+science", "0123_abcd science"))
````

These kinds of matching, even with relatively simple starting points, can lead to some interesting applications:

**Aside:** One thing you may notice is the `r""` format of the regular expressions (quotes with an 'r' preceding them).  You can actually use any string as a regular expression, but the `r` expressions are quite handy for the following reason.  In a typical Python string, backslash characters denote escaped characters, so for instance `"\\"` really just encodes a single backslash.  But backslashes are _also_ used within regular expressions.  So if we want the regular expression `\\` represented as a string (that is, match a single backslash), we'd need to use the string `"\\\\"`.  This gets really tedious quickly.  So the `r""` notation just _ignores_ any handling of handling of backslashes, and thus makes inputing regular expressions much simpler.

````python
print("\\")
print(r"\\")
````

### Grouping

Beyond the ability to just match strings, regular expressions also let you easily find specific sub-elements of the matched strings.  The basic syntax is the following: if we want to "remember" different portions of the matched expression, we just surround those portions of the regular expression in parentheses.  For example, the regular expression `r"(\w+)\s([Ss]cience)"` would store whatever element is matched to the `\w+` and `[Ss]cience` portions in the `groups()` object in the returned match.

````python
match = re.search(r"(\w+)\s([Ss]cience)", text)
print(match.groups())
````

The `.group(i)` notation also lets you easily find just individual groups, `.group(0)` being the entire text.

````python
match = re.search(r"(\w+)\s([Ss]cience)", text)
print(match.group(0))
print(match.group(1))
print(match.group(2))
````

### Substitutions



Regular expression can also be used to automatically substitute one expression for another within the string.  This is done using the `re.sub()` call.  This returns a string with (all) the instances of the first regular expression replaced with the second one.  For example, to replace all the occurrences of "data science" with "data schmience", we could use the following code:

````python
print(re.sub(r"data science", r"data schmience", text))
````

Where this gets really powerful is when we use groups in the first regular expression.  These groups can then be backreferenced using the `\1`, `\2`, etc notation in the second one (more generally, you can actually use these backreferencs within a single regular expression too, outside the context of substitutions, but we won't cover that here).  So if we have the regular expression `r"(\w+) ([Ss])cience"` to match "*something* science" (where science can be capitalized or not), we would replace it with the string "*something* schmience", keeping the *something* the same, and keeping the capitalization of science the same, using the code:

````python
print(re.sub(r"(\w+) ([Ss])cience", r"\1 \2chmience", text))
print(re.sub(r"(\w+) ([Ss])cience", r"\1 \2chmience", "Life Science"))
````

### Miscellaneous items



There are a few last points that we'll discuss here.

**Order of operations.** The first point comes in regard to the order of operations for regular expressions.  The `|` character in regular expressions is like an "or" clause, the regular expression should can match the regular expression to the left or to the right of the character.  For example, the regular expression `r"abc|def"` would match the string "abc" or "def".

````python
print(re.match(r"abc|def", "abc"))
print(re.match(r"abc|def", "def"))
````

But what if we want to match the string "ab*(c or d)*ef"?  We can capture this in a regular expression by parentheses around the portion we want to give a higher order of operations.

````python
print(re.match(r"abc|def", "abdef"))
print(re.match(r"ab(c|d)ef", "abdef"))
````

But, of course, since we also use the parentheses for specifying groups, this means that we will be creating a group for the *c or d* character here.  While it's probably not the end of the world to create a few groups you don't need, in the case that you _don't_ want to create this group, you can use the notation `r"ab(?:c|d)ef"`.  There is no real logic to this notation that I can see, it just happens to be shorthand for "use these parentheses to manage order of operations but don't create a group."  Regular expressions are full of fun things like this, and while you likely don't need to commit this to memory, it's handy to remember that there are situations like this.

````python
print(re.match(r"ab(c|d)ef", "abdef").groups())
print(re.match(r"ab(?:c|d)ef", "abdef").groups())
````

**Greedy and lazy matching** The second point is something that you likely _will_ need to know.  This is the fact that, by default, regular expressions will always match _the longest possible string_.  So suppose we have the regular expression `r"<.*>"` (a less-than, followed by any number of any character, followed by a greater-than).  As you might imagine, this sort of expression will come up if (and for this class, when) you're writing a parser for a format like XML.  If we try to match the string "&lt;tag>hello&lt;/tag>", then the regular expression will match the _entire_ string (since the entire string is a less-than, followed by some number of characters, followed by a greater-than); it will not just match the "&lt;tag>" part of the string.  This is known as "greedy" matching.

````python
print(re.match(r"<.*>", "<tag>hello</tag>"))
````

In the case where you don't want this, but would instead want you match the _smallest_ possible string, you can use the alternative regular expression `r"<.*?>"`.  The `*?` notation indicates "lazy" matching, that you want to match any number of characters possible, but would prefer the _smallest_ possible match.  This will then match just the "&lt;tag>" string.

````python
print(re.match(r"<.*?>", "<tag>hello</tag>"))
````
